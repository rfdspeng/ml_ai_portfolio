```python
from qdrant_client import QdrantClient
import os
```

```python
vectordb = QdrantClient(os.environ["QDRANT_URL"], api_key=os.environ["QDRANT_API_KEY"])
```

```python
collection_name = "best-book-ever"
```

```python
# vectordb.create_snapshot(collection_name=collection_name)
# vectordb.delete_snapshot(collection_name=collection_name, snapshot_name=snapshots[0].name)
snapshots = vectordb.list_snapshots(collection_name=collection_name)
snapshots
```

```python
collection_info = vectordb.get_collection(collection_name=collection_name)
collection_info
```

```python
points = list(range(collection_info.points_count))
points
```

```python
vectordb.delete_payload(
    collection_name=collection_name,
    keys=["descriptor"],
    points=points,
)
```

```python
import pandas as pd

df = pd.read_pickle("bbe_12-05-2024-2236.pkl")
```

```python
descriptions = df["description"].to_list()
genres = df["genres"].to_list()
```

```python
print(len(descriptions))
print(len(genres))
```

```python
for point in points[689:]:
    vectordb.set_payload(
        collection_name=collection_name,
        payload={
            "description": descriptions[point],
            "genres": genres[point],
        },
        points=[point]
    )
    print(f"Updated point {point}")
    # payload={
    #     "description": descriptions[point],
    #     "genres": genres[point],
    # }
    # print(payload["description"])
    # print(payload["genres"])
    # print("---")
```
