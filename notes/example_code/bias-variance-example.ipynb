```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
for i in range(5):
    X, y, coef = make_regression(n_samples=10, n_features=1, n_informative=1, bias=0, noise=30, coef=True, random_state=i)
    print(coef)
```

```python
f, ax = plt.subplots()
reg = LinearRegression()
ord = 3
for i in range(3):
    np.random.seed(i)
    e = np.random.normal(scale=2, size=(10,))
    X = np.arange(0, 10, 1)
    y = 0.5*(X ** 3) - (X ** 2) + 2*X + 1 + e
    ax.scatter(X, y)

    reg.fit(X[:, np.newaxis] ** list(range(ord+1)), y)
    X_model = np.linspace(np.min(X), np.max(X), num=100)[:, np.newaxis]
    y_pred = reg.predict(X_model ** list(range(ord+1)))
    ax.plot(X_model, y_pred)

# ax.set_ylim([-5, 30])
ax.grid()
```
