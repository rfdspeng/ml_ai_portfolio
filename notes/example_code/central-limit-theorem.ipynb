```python
X = np.floor(np.random.uniform(low=1, high=7, size=(10000,)))
```

```python
print(X.mean())
print(X.std())
```

```python
ax = sns.histplot(X, discrete=True, stat="probability")
ax.set_xlabel("Die Roll")
ax.set_title("Distribution of 10000 Die Rolls")
```

```python
Xbar = np.array([np.random.choice(X, size=(50,)).mean() for i in range(10000)])
```

```python
print(Xbar.mean())
print(Xbar.std())
```

```python
ax = sns.histplot(Xbar, binwidth=0.1, kde=True, stat="probability")
ax.set_xlabel("Average Die Roll (50 Rolls)")
ax.set_title("Sampling Distribution of Die Roll Average")
ax.grid()
```
