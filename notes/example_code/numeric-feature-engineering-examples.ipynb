```python
import numpy as np
from numpy.random import uniform
import seaborn as sns

raw = uniform(0, 100, size=1000)
binned = np.floor(raw/10)

g = sns.relplot(x=raw, y=binned, kind="scatter")
g.ax.set_xlabel("Raw Data")
g.ax.set_ylabel("Bin")
g.ax.grid()
g.ax.set_xticks(np.arange(0, 101, 10))
g.ax.set_yticks(np.arange(0, 11, 1))
```

```python
raw = uniform(1, 10000, size=10000)
binned = np.floor(np.log10(raw))

g = sns.relplot(x=raw, y=binned, kind="scatter")
g.ax.set_xscale("log")
g.ax.set_xlabel("Raw Data")
g.ax.set_ylabel("Bin")
g.ax.grid()
# g.ax.set_xticks(np.arange(0, 101, 10))
g.ax.set_yticks(np.arange(0, 4, 1))
```
