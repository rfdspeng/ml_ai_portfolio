```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

slope = 4
y_int = 60

# x = np.arange(0, 11, 1)
x = np.random.rand(50)*10
y_n = slope*x + y_int + (np.random.rand(x.size)*2-1)*7
y_n[y_n > 100] = 100
y = slope*x + y_int

fig = plt.figure(figsize=(8,5), dpi=300)
plt.scatter(x, y_n, s=mpl.rcParams['lines.markersize'] ** 2 * 5, label="Data Points")
plt.plot(x, y, label="Predictive Model", color="k", linewidth=5)
plt.title("Regression", fontsize=30)
plt.xlabel("x = Hours Studied", fontsize=20)
plt.ylabel("y = Grade", fontsize=20)
plt.legend(loc="best")
# plt.xlim([0, max(x)])
# plt.ylim([60, max(y)])
plt.grid()
```

```python
from sklearn.linear_model import SGDRegressor

fig = plt.figure(figsize=(8,5), dpi=300)

sgd_reg = SGDRegressor(random_state=0)

for i in range(201):
    sgd_reg.partial_fit(x.reshape(-1,1), y_n)
    if i in [0, 20, 100, 200]:
        x_plot = np.arange(0, 11, 1)
        y_plot = x_plot*sgd_reg.coef_ + sgd_reg.intercept_
        plt.plot(x_plot, y_plot, label=f"Model, {i}-th Iteration", linewidth=3)

plt.scatter(x, y_n, s=mpl.rcParams['lines.markersize'] ** 2 * 5, label="Data Points", c="k")
plt.title("Model Training", fontsize=30)
plt.xlabel("x = Hours Studied", fontsize=20)
plt.ylabel("y = Grade", fontsize=20)
plt.legend(loc="best")
plt.grid()

print(sgd_reg.coef_)
print(sgd_reg.intercept_)
```

```python
import matplotlib.pyplot as plt
import numpy as np

slope = 1
y_int = -5

x_n = np.random.rand(50)*10
y_n = slope*x_n + y_int
x = np.arange(0,10,0.1)
y = slope*x + y_int
y_n = 1/(1+np.exp(-y_n))
y = 1/(1+np.exp(-y))

threshold = 0.5
y_n = y_n + (np.random.rand(x_n.size)*1-0.5) >= threshold

fig = plt.figure(figsize=(8,5))
plt.scatter(x_n, y_n, s=mpl.rcParams['lines.markersize'] ** 2 * 5, label="Data Points")
plt.plot(x, y, label="Predictive Model", color="k", linewidth=5)
plt.title("Classification", fontsize=30)
plt.xlabel("x = Hours Studied", fontsize=20)
plt.ylabel("y = Passing Grade", fontsize=20)
plt.legend(loc="best")
# plt.xlim([0, max(x)])
# plt.ylim([0, max(y)])
plt.grid()
```
