```python
import torch
from torch.optim.lr_scheduler import LambdaLR
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_epochs = 10
lr_init = 1.6e-3

linear_step = lr_init/num_epochs

lr_linear = lr_init - linear_step*(np.array(range(0,num_epochs+1)))
```

```python
plt.figure()
x_lin = list(range(1,num_epochs+2))
y_lin = lr_linear
plt.plot(x_lin, y_lin, marker='o')
plt.xlim([1, num_epochs+1])
plt.ylim([0, np.max(y_lin)])
plt.grid()
```

```python
num_train = 1040 # number of training samples
ebs = 8 # effective batch size
step_size = ebs/num_train # step size in units of epoch
```

```python
gamma = 0.999
num_epochs = 30
lr_init = lr_init
steps = np.arange(0, num_epochs+1, step_size)
powers = np.array(range(len(steps)))
lr_step = lr_init*np.power(gamma*np.ones(len(steps)), powers)
```

```python
plt.figure()
x_step = steps+1
y_step = lr_step
plt.plot(x_lin, y_lin, marker='o')
plt.plot(x_step, y_step)
plt.xlim([1, num_epochs+1])
plt.ylim([0, np.max(y_lin)])
plt.grid()
```
